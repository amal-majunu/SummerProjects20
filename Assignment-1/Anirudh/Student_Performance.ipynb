Adding Missing Roll Numbers

```
# Adding Student with Roll_No 1
new_row = pd.DataFrame({'Roll': 1}, index =[0]) 
df = pd.concat([new_row, df]).reset_index(drop = True) 

# Filling in Missing Roll Numbers
df['Roll'] = df.index + 1

# Setting Index as 'Roll'
df.set_index('Roll', inplace = True)

df.head()
```

Treating Invalid Entries


```
# Setting invalid Entries to NaN
df.loc[df['CT-1 (Marks: 20)']  == 'CT-1 (Marks: 20)'] = np.NaN

# Set type of all columns as 'float'
df = df.astype(float)

pd.set_option("max_rows", 15)

# Set missing class test marks as Median
df.loc[pd.isnull(df['CT-1 (Marks: 20)']), 'CT-1 (Marks: 20)'] = df.loc[pd.notnull(df['CT-1 (Marks: 20)']), 'CT-1 (Marks: 20)'].median()
df.loc[pd.isnull(df['CT-2 (Marks: 20)']), 'CT-2 (Marks: 20)'] = df.loc[pd.notnull(df['CT-2 (Marks: 20)']), 'CT-2 (Marks: 20)'].median()
df.loc[pd.isnull(df['CT-3 (Marks: 20)']), 'CT-3 (Marks: 20)'] = df.loc[pd.notnull(df['CT-3 (Marks: 20)']), 'CT-3 (Marks: 20)'].median()
df.loc[pd.isnull(df['CT-4 (Marks: 20)']), 'CT-4 (Marks: 20)'] = df.loc[pd.notnull(df['CT-4 (Marks: 20)']), 'CT-4 (Marks: 20)'].median()

# Set Overall Class test marks best of 3 CT (Setting all values in case some values were not calculated correctly)
df['Class Test (Total of best 3, Marks: 60)'] = df[['CT-1 (Marks: 20)', 'CT-2 (Marks: 20)', 'CT-3 (Marks: 20)', 'CT-4 (Marks: 20)']].sum(axis = 1, skipna = True) 
df['Class Test (Total of best 3, Marks: 60)'] -= df[['CT-1 (Marks: 20)', 'CT-2 (Marks: 20)', 'CT-3 (Marks: 20)', 'CT-4 (Marks: 20)']].min(axis = 1, skipna = True)

# Set missing Attendence marks
df.loc[pd.isnull(df['Attendance (Marks: 15)']), 'Attendance (Marks: 15)'] = df.loc[pd.notnull(df['Attendance (Marks: 15)']), 'Attendance (Marks: 15)'].median()

# Setting missing Observation marks
df.loc[pd.isnull(df['Observation (Marks: 15)']), 'Observation (Marks: 15)'] = df.loc[pd.notnull(df['Observation (Marks: 15)']), 'Observation (Marks: 15)'].median()

# Setting all Total marks (90) (setting all values in case they were not calculate correctly)
df['Total (Marks: 90)'] = df[['Class Test (Total of best 3, Marks: 60)', 'Attendance (Marks: 15)', 'Observation (Marks: 15)']].sum(axis = 1)

# Renaming Exam Mark and Exam Mark .1 to suitable names
df = df.rename(columns = {'Exam Mark (105)' : 'Exam-1 (Marks: 105)', 'Exam Mark (105).1' : 'Exam-2 (Marks: 105)', 'Mark total 300' : 'Total (Marks: 300)'})

# Setting missing Exam-1 and Exam-2 marks
df.loc[pd.isnull(df['Exam-1 (Marks: 105)']), 'Exam-1 (Marks: 105)'] = df.loc[pd.notnull(df['Exam-1 (Marks: 105)']), 'Exam-1 (Marks: 105)'].median()
df.loc[pd.isnull(df['Exam-2 (Marks: 105)']), 'Exam-2 (Marks: 105)'] = df.loc[pd.notnull(df['Exam-2 (Marks: 105)']), 'Exam-2 (Marks: 105)'].median()

# Setting Total marks (300) (Setting all incase they where not calculated correctly)
df['Total (Marks: 300)'] = df[['Exam-1 (Marks: 105)', 'Exam-2 (Marks: 105)', 'Total (Marks: 90)']].sum(axis = 1)

# Adding Attendance Percentage Column
df['Attendance Percentage'] = df['Attendance (Marks: 15)'] / 15 * 100

df
```

```
# Function to Group Students based on performance in individual tests
def judge_performance(test):
  label = test + " Performance"
  location = df.columns.get_loc(test)
  df.insert(location + 1, label, pd.cut(np.array(df[test]), 4, labels = ['very bad', 'bad', ' good', 'very good']))

# Judging the performance of Students in all Test
for column in df.columns:
  if(column != 'Attendance (Marks: 15)' and column != 'Observation (Marks: 15)' and column != 'Attendance Percentage'):
    judge_performance(column)

df
```
